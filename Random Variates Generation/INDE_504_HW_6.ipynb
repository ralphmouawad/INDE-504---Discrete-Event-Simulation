**HW 6 - Generating Random Variates**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

Exercise 8.4 - Generate from triang(1, 10, 4)

```python
X = []
m1 = 1/3
for _ in range(1000):
  U = random.uniform(0,1)
  if U < m1:
    y = np.sqrt(m1*U)
  else:
    y = np.sqrt(np.square(1-U) / 3)
  x = 1 + 9*y
  X.append(x)
```

```python
plt.hist(X, bins=30)
plt.xlabel('X')
plt.ylabel('Frequency')
plt.title('Histogram of X')
plt.show()
```

Ex. 8.5 - Generating from a given CDF

```python
X = []
for _ in range(1000):
  U = random.uniform(0,1)
  if 0 < U <= 0.5:
    X.append(6*U-3)
  elif 1/2 < U <= 1:
    X.append(4*np.sqrt(2*U-1))
```

```python
plt.hist(X, bins=30)
plt.xlabel('X')
plt.ylabel('Frequency')
plt.title('Histogram of X')
plt.show()
```

Ex. 8.29 - Generating from NSPP

```python
t=0
i=1
T=[t]


def compute_lambda(t):
  if 0<=t<1:
    l = 30
  if 1<=t<2:
    l = 45
  if 2<=t<=4:
    l = 60
  return l


while len(T) < 100:
  t = T[-1]
  U1 = random.uniform(0,1)
  U2 = random.uniform(0, 1)
  t = t - 1/45 * math.log10(U1)
  l = compute_lambda(t)
  if U2 <= l/45:
    T.append(t)
  else:
    continue


```

```python
plt.hist(T, bins=30)
plt.xlabel('X')
plt.ylabel('Frequency')
plt.title('Histogram of X')
plt.show()
```
